```python
import matplotlib.pyplot as plt
import numpy as np
```

**<font size="5">Bar Plot</font>**

```python
P_lang=['Java','Python','PHP','JS','C#','C++']
pop=[22.2,17.6,8.8,8,7.7,6.7]
```

```python
plt.bar(P_lang,pop,width=0.5)
plt.title('Bar Plot')
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
```

**<font size="5">Multiple Bars</font>**

```python
N = 5
ind = np.arange(N)  # the x locations for the groups
width = 0.27       # the width of the bars

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

men = [22,30,35,35,26]
rects1 = ax.bar(ind,men, width, color='r')
women = [25,32,30,35,29]
rects2 = ax.bar(ind+width,women, width, color='g')

ax.set_ylabel('Scores')
ax.legend( (rects1[0], rects2[0]), ('men', 'women') )

def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2., 1.0015*h, '%d'%int(h),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```

**<font size="5">Stack bar charts</font>**

```python
N = 5
men = (22,30,35,35,26)
women = (25,32,30,35,29)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, men, width)
p2 = plt.bar(ind, women, width,
             bottom=men)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))

plt.show()
```

**<font size="5">Pie Chart</font>**

```python
P_lang=['Java','Python','PHP','JS','C#','C++']
pop=[22.2,17.6,8.8,8,7.7,6.7]
```

```python
plt.pie(pop,explode=(0,0.1,0,0,0,0), labels=P_lang, autopct='%1.1f%%',shadow=True, startangle=90)
plt.show()
```
